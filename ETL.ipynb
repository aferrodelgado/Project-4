```python
# Import Libraries
import pandas as pd
```

```python
# Read in CSV
file_path = pd.read_csv("Fraudulent_E-Commerce_Transaction_Data_2.csv")

# Create Dataframe
transaction_df = pd.DataFrame(file_path)
transaction_df.head()
```

```python
# Drop unnessecary colummns
transaction_df = transaction_df.drop(columns=['Transaction ID', 'Customer ID', 'IP Address', 'Shipping Address', 'Billing Address'])
transaction_df.head()
```

```python
# Convert columns to datetime
transaction_df['Transaction Date'] = pd.to_datetime(transaction_df['Transaction Date'])

transaction_df['year'] = transaction_df['Transaction Date'].dt.year
transaction_df['month'] = transaction_df['Transaction Date'].dt.month
transaction_df['day'] = transaction_df['Transaction Date'].dt.day

transaction_df.head()
```

```python
# Rename columns
transaction_df = transaction_df.rename(columns={
    'Transaction Amount': 'trxn_amt', 
    'Transaction Date': 'trxn_date', 
    'Payment Method': 'pmt_method', 
    'Product Category': 'product_cat', 
    'Quantity': 'qty',
    'Customer Age': 'cust_age',
    'Customer Location': 'cust_loc',
    'Device Used': 'cust_device',
    'Is Fraudulent': 'fraudulent',
    'Account Age Days': 'acct_age_days',
    'Transaction Hour': 'trxn_hour'
})
transaction_df.head()
```

```python
# Drop 'trxn_date' column
transaction_df = transaction_df.drop(columns=['trxn_date'])
```

## Visualization csv

```python
# Save dataframe as csv for visualizations
transaction_df.to_csv('visualization_transaction_df.csv', index=False, header=True)
```

## Use Standard Scaler and Encode non-integer values for modeling

```python
# import libraries
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
# Encode non_integer values
encoded_dummies = pd.get_dummies(transaction_df, columns=['pmt_method', 'product_cat', 'cust_device'])

# display data
encoded_dummies.head()
```

```python
# Scale numeric data
scaled_data = StandardScaler().fit_transform(encoded_dummies[['trxn_amt', 'acct_age_days']])

scaled_data
```

```python
# create dataframe from scaled data and replace old columns with scaled data
transaction_scaled_data = pd.DataFrame(scaled_data, columns=['trxn_amt', 'acct_age_days'])

# Replace old columns in dummies table
encoded_dummies['trxn_amt'] = transaction_scaled_data['trxn_amt']
encoded_dummies['acct_age_days'] = transaction_scaled_data['acct_age_days']

# Review dataframe
encoded_dummies.head()
```

```python
# Rename dataframe and drop 'cust_loc' column
preprocessed_df = encoded_dummies

preprocessed_df = preprocessed_df.drop(columns=['cust_loc'])
```

## Modeling csv

```python
# Save csv for modeling
preprocessed_df.to_csv('modeling_df.csv', index=False, header=True)
```
